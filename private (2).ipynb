```python
import numpy as np

def weighted_sum(inputs, weights, bias):
    return np.dot(inputs, weights) + bias

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def forward_propagation(inputs, weights, bias):
    z = weighted_sum(inputs, weights, bias)
    
    output = sigmoid(z)
    
    return output, z

def backward_propagation(inputs, weights, bias, output, target, z, learning_rate):
    error = output - target
    
    d_output = error * sigmoid_derivative(z)
    
    d_weights = np.dot(inputs.T, d_output)
    d_bias = np.sum(d_output)
    
    weights -= learning_rate * d_weights
    bias -= learning_rate * d_bias
    
    return weights, bias, d_weights, d_bias, error

inputs = np.array([1.5, -2.0, 0.8, 3.0])
weights = np.array([0.4, -0.5, 0.6, 0.3])
bias = -1.2
target = 1  
learning_rate = 0.01  

output, z = forward_propagation(inputs, weights, bias)

weights, bias, d_weights, d_bias, error = backward_propagation(inputs, weights, bias, output, target, z, learning_rate)

# Print results
print("Updated weights:", weights)
print("Updated bias:", bias)
print("Weight gradients:", d_weights)
print("Bias gradient:", d_bias)
print("Error:", error)
```

```python
import numpy as np
import pandas as pd

def weighted_sum(inputs, weights, bias):
    return np.dot(inputs, weights) + bias

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(output):
    return output * (1 - output)

def forward_propagation(X, weights, bias):
    z = weighted_sum(X, weights, bias)
    output = sigmoid(z)
    return output

def backward_propagation(X, y, output, weights, bias, learning_rate=0.01):
    error = output - y
    d_weights = np.dot(X.T, error * sigmoid_derivative(output))
    d_bias = np.sum(error * sigmoid_derivative(output))
    weights -= learning_rate * d_weights
    bias -= learning_rate * d_bias
    return weights, bias, error

df = pd.read_csv('x.csv')
print("Dataset:")
print(df.head())

X = df.drop('target_column', axis=1).values
y = df['target_column'].values

weights = np.random.randn(X.shape[1])
bias = np.random.randn()

epochs = 1000
learning_rate = 0.01

for epoch in range(epochs):
    output = forward_propagation(X, weights, bias)
    weights, bias, error = backward_propagation(X, y, output, weights, bias, learning_rate)
    if epoch % 100 == 0:
        loss = np.mean(error**2)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")

print("\nFinal weights:", weights)
print("Final bias:", bias)

predictions = forward_propagation(X, weights, bias)
print("\nPredictions after training:")
print(predictions)
```
